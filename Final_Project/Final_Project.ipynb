```python
def evaluate_preds(train_true_values, train_pred_values, test_true_values, test_pred_values):
    print("Train R2:\t" + str(round(r2(train_true_values, train_pred_values), 3)))
    print("Test R2:\t" + str(round(r2(test_true_values, test_pred_values), 3)))
    
    plt.figure(figsize=(18,10))
    
    plt.subplot(121)
    sns.scatterplot(x=train_pred_values, y=train_true_values)
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('Train sample prediction')
    
    plt.subplot(122)
    sns.scatterplot(x=test_pred_values, y=test_true_values)
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('Test sample prediction')

    plt.show()
```

### Функция, которая заменяет значения в столбце ниже х квантиля и выше (1-х) квантиля на х и (1-х) квантили соответственно
### (обработка выбросов)

```python
def ProcessColumn(df, column_name, quantile_limit):
    df1 = df.copy()
    lower = df[column_name].quantile(quantile_limit)
    upper = df[column_name].quantile(1 - quantile_limit)
    df1[column_name] = df1[column_name].where(df1[column_name] > lower, lower, axis=0)
    df1[column_name] = df1[column_name].where(df1[column_name] < upper, upper, axis=0)
    return df1
```

### Если значение HouseFloor равно 0, то приравняем его к соответстующему значению Floor

```python
df_train['HouseFloor'] = df_train['HouseFloor'].where(df_train['HouseFloor'] > 0, df_train['Floor'], axis=0)
```

### Поскольку, в сете есть много значений этажей равных 'Этажность + 1', предположу, что это чердаки и оставлю их. В остальных случаях, если этаж больше этажности, то приравняю его к значению среднего этажа для данного дома

```python
df_train['Floor'] = df_train['Floor'].where(df_train['Floor'] < df_train['HouseFloor'] + 2, df_train['HouseFloor'] // 2 + 1, axis=0)
```

## Обработаем выбросы в KitchenSquare

### Найдем среднее значение отношения KitchenSquare к LifeSquare для всех домов

```python
ratio_kitchen = (df_train['KitchenSquare'] / df_train['LifeSquare']).mean()
```

### Значения KitchenSquare. большие, чем соответствующие значения LifeSquare, а также, значение меньше 3, считаем выбросами и заменяем на (ratio_kitchen * LifeSquare)

```python
df_train['KitchenSquare'] = df_train['KitchenSquare'].where(df_train['KitchenSquare'] <= df_train['LifeSquare'], df_train['LifeSquare'] * ratio_kitchen, axis = 0)
```

```python
df_train['KitchenSquare'] = df_train['KitchenSquare'].where(df_train['KitchenSquare'] >= 3, df_train['LifeSquare'] * ratio_kitchen, axis = 0)
```

## Обработаем выбросы в Rooms

```python
df_train.groupby('Rooms').count()
```

### Будем считать значения больше 5 комнат выбросами. Найдём среднее отношение числа комнат к жилой площади и исходя из него заменим выбросы

```python
ratio_rooms = (df_train['Rooms'] / df_train['LifeSquare']).mean()
```

```python
df_train['Rooms'] = df_train['Rooms'].where(df_train['Rooms'] <= 5, round(ratio_rooms * df_train['LifeSquare'], 0), axis=0)
```
